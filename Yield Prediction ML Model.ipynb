```sql
SELECT
  `Corn Yield`,
  `Soybean Yield`,
  `County Name`,
  `Year`,
  `MaxT`,
  `MinT`,
  `Precipitation`,
  `Cooling`,
  `Heating`,
  `AvgT`,
  `PDSI`
FROM workspace.default.master
WHERE
  CAST(`PDSI` AS DOUBLE) BETWEEN CAST(:PDSI AS DOUBLE) - 1 AND CAST(:PDSI AS DOUBLE) + 1 AND
  CAST(`MaxT` AS DOUBLE) BETWEEN CAST(:MaxT AS DOUBLE) - 1 AND CAST(:MaxT AS DOUBLE) + 1 AND
  CAST(`MinT` AS DOUBLE) BETWEEN CAST(:MinT AS DOUBLE) - 1 AND CAST(:MinT AS DOUBLE) + 1 AND
  CAST(`Precipitation` AS DOUBLE) BETWEEN CAST(:Precipitation AS DOUBLE) - 1 AND CAST(:Precipitation AS DOUBLE) + 1 AND
  CAST(`Cooling` AS DOUBLE) BETWEEN CAST(:Cooling AS DOUBLE) - 50 AND CAST(:Cooling AS DOUBLE) + 50 AND
  CAST(`Heating` AS DOUBLE) BETWEEN CAST(:Heating AS DOUBLE) - 50 AND CAST(:Heating AS DOUBLE) + 50 AND
  CAST(`AvgT` AS DOUBLE) BETWEEN CAST(:AvgT AS DOUBLE) - 1 AND CAST(:AvgT AS DOUBLE) + 1
ORDER BY ABS(CAST(`Corn Yield` AS DOUBLE) - (SELECT AVG(`Corn Yield`) FROM workspace.default.master)), ABS(CAST(`Soybean Yield` AS DOUBLE) - (SELECT AVG(`Soybean Yield`) FROM workspace.default.master))
LIMIT 3
```

```sql
%python
import mlflow
import mlflow.spark
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor, LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# 1. Load the data
master_df = spark.table('workspace.default.master')

features = ["MaxT", "MinT", "Precipitation", "Cooling", "Heating", "AvgT"]
targets = ["Corn Yield", "Soybean Yield"]

UC_VOLUME_PATH = '/Volumes/workspace/default/databrickshackathon'

for target in targets:
    print(f"\n--- {target} ---\n")
    # 2. Feature vector assembly
    assembler = VectorAssembler(inputCols=features, outputCol="features")
    data = assembler.transform(master_df).select(col("features"), col(target).alias("label"))
    # 3. Drop missing target rows
    data = data.na.drop(subset=["label"])
    # 4. Train/test split
    train, test = data.randomSplit([0.8, 0.2], seed=42)

    for model_type, regressor in {
        'rf': RandomForestRegressor(featuresCol='features', labelCol='label', seed=42),
        'lr': LinearRegression(featuresCol='features', labelCol='label')
    }.items():
        with mlflow.start_run(run_name=f"{model_type}_{target.replace(' ', '_')}"):
            pipeline = Pipeline(stages=[regressor])
            mdl = pipeline.fit(train)
            preds = mdl.transform(test)
            predictions = preds.select("label", "prediction").toPandas()
            mlflow.log_param("model_type", model_type)
            mlflow.log_param("target", target)
            mlflow.log_param("features", ",".join(features))
            r2 = r2_score(predictions['label'], predictions['prediction'])
            rmse = np.sqrt(mean_squared_error(predictions['label'], predictions['prediction']))
            mlflow.log_metric("r2", r2)
            mlflow.log_metric("rmse", rmse)
            mlflow.spark.log_model(mdl, f"{model_type}_{target.replace(' ', '_')}_model", dfs_tmpdir=UC_VOLUME_PATH)
            print(f"{model_type.upper()} | R2: {r2:.3f} | RMSE: {rmse:.3f}")
print("\nAll modeling runs and metrics are tracked in MLflow (click the Experiment icon on the left sidebar).\n")
```

```sql
%python
import mlflow
import mlflow.spark
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

UC_VOLUME_PATH = '/Volumes/workspace/default/databrickshackathon'
master_df = spark.table('workspace.default.master')
features = ["MaxT", "MinT", "Precipitation", "Cooling", "Heating", "AvgT"]
targets = ["Corn Yield", "Soybean Yield"]

for target in targets:
    assembler = VectorAssembler(inputCols=features, outputCol="features")
    data = assembler.transform(master_df).select("features", target)
    data = data.na.drop(subset=[target])
    train, test = data.randomSplit([0.8, 0.2], seed=42)

    search_space = {
        'numTrees': hp.quniform('numTrees', 20, 120, 1),
        'maxDepth': hp.quniform('maxDepth', 2, 10, 1),
        'maxBins': hp.quniform('maxBins', 32, 128, 1)
    }

    def objective(params):
        with mlflow.start_run(run_name=f"hyperopt_rf_{target.replace(' ', '_')}"):
            rf = RandomForestRegressor(featuresCol="features", labelCol=target,
                                      numTrees=int(params['numTrees']),
                                      maxDepth=int(params['maxDepth']),
                                      maxBins=int(params['maxBins']),
                                      seed=42)
            pipeline = Pipeline(stages=[rf])
            mdl = pipeline.fit(train)
            preds = mdl.transform(test)
            predictions = preds.select(target, "prediction").toPandas()
            r2 = r2_score(predictions[target], predictions["prediction"])
            rmse = np.sqrt(mean_squared_error(predictions[target], predictions["prediction"]))
            mlflow.log_params(params)
            mlflow.log_metric("r2", r2)
            mlflow.log_metric("rmse", rmse)
            mlflow.spark.log_model(mdl, f"rf_{target.replace(' ', '_')}_model", dfs_tmpdir=UC_VOLUME_PATH)
            return {'loss': -r2, 'status': STATUS_OK, 'metrics': {'r2': r2, 'rmse': rmse}, 'model': mdl}

    trials = Trials()
    best = fmin(fn=objective,
                space=search_space,
                algo=tpe.suggest,
                max_evals=15,
                trials=trials)

    best_trial = max(trials.trials, key=lambda t: t['result']['metrics']['r2'])
    result = best_trial['result']['metrics']
    print(f"\nBest Random Forest for {target}: R2={result['r2']:.3f}, RMSE={result['rmse']:.3f}, Params={best}")
print("\nAll Hyperopt tuning results and models are tracked in MLflow.")
```
